## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
My chosen lane is identifying webpages that are likely to decline in search performance.

This is a Classification problem because the model predicts whether a webpage is declining or not declining.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
The target is `is_declining_label`.

This target identifies whether a webpage is declining in search performance. It is created from the `trend_direction` column, where pages with the value `down` are labeled as 1 (declining) and all other pages are labeled as 0 (not declining).
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
The success metric for this project is Precision@50.

Precision@50 measures how many of the top 50 webpages selected by the model are actually declining. A higher Precision@50 means the model is better at identifying the pages that should be reviewed first.
```
